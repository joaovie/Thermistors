```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#Steinhart–Hart equation relative to Temperature
def Temperature(R,B,T0,R0):
    temp=R/R0 #(R/Ro)
    temp=np.log(temp) #ln(R/Ro)
    temp /= B #1/B * ln(R/Ro)
    temp += 1 / (T0 + 273.15) # + (1/To)
    temp= 1/temp #invert
    temp -= 273.15 # Convert absolute temp to C
    return temp
```

```python
#Steinhart–Hart equation relative to Resistance
def Resistance(T,B,T0,R0):
    Res=1/(T+ 273.15)
    Res-=1/(T0 + 273.15)
    Res *=B
    Res= math.exp(Res)
    Res*=R0
    return Res
```

Ohmite TX04F103F3380ER NTC

```python
T0=25 #C
#Beta parameters and nominal resistance
B_Ohmite=3380 #K
R0_Ohmite=10000 #Ohm


R=np.linspace(1,12000,3000)

plt.plot(R/1000, Temperature(R,B_Ohmite,T0,R0_Ohmite), color='red')
plt.ylim([20,80])
plt.xlim([1.5,12])
plt.title('Ohmite TX04F103F3380ER Resistance Variation\n with Temperature')
plt.xlabel("R (kOhms)")
plt.ylabel("Temperature (°C)")
plt.grid()

plt.show()
```

```python

R_Delta=Resistance(25,B_Ohmite,T0,R0_Ohmite)-Resistance(24,B_Ohmite,T0,R0_Ohmite)
print('With the increase of 1 °C there is a decrease of ',abs(R_Delta), 'Ohms')
```
